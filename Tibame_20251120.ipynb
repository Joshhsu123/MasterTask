````python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
print(is_leap_year(2010))
````
