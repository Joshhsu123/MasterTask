````python
def is_leap_year(year):
    ans = False
    if year % 400 == 0:
        ans = True
    elif year % 100 == 0:
        ans = False
    elif year % 4 == 0:
        ans = True
    else:
        ans = False
    return ans
print(is_leap_year(2010))
````

````python
#預設答案是合理, 再不合理的地方設為False
def is_date_valid(date_checked):
    #date_checked = 20251301
    day = date_checked % 100
    remain = date_checked // 100
    month = remain % 100
    year = remain // 100
    ans = True
    if year > 0:
        if 1 <= month <= 12:
            day_limit_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_limit = day_limit_per_month[month-1]
            if month == 2 and is_leap_year(year) == True:
                day_limit = day_limit + 1
                if 1 <= day <= day_limit:
                    print("日期合理")
                    ans = True
                else:
                    print("日期不合理")
                    ans = False
            else:
                print("月份不合理")
                ans = False
    else:
        print("年份不合理")
        ans = False
    return ans
````

````python
import datetime as dt

def fully_squared_2(start_y=2024, start_m=10, start_d=1, max_count=10):
    ans = []
    pat = "%Y%m%d"
    start = dt.datetime(start_y, start_m, start_d)
    count = 0
    while True:
        #轉成String
        start_str = start.strftime(pat)
        #轉成整數
        start_int = int(start_str)
        start_root = start_int ** 0.5
        # 四捨五入-原始 < 某個數值 我就認為依樣
        # 3.9999999999 == 4 python要差到很小才會一樣
        dif = abs(round(start_root) - start_root)
        if dif < 0.000001:
            print(start)
            ans.append(start_int)
            count = count + 1
        if count == max_count:
            break
        start = start + dt.timedelta(days=1)
    return ans
fully_squared_2()
````
